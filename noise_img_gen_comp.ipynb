```python
from random import randint
```

```python
import matplotlib.pyplot as plt
```

```python
def py_noiser(hres, vres):
    return [[[randint(0,255) for i in range(3)] for i in range (hres)] for i in range(vres)] #must be 0-255
```

```python
%timeit py_noiser(1920,1080)
```

```python
img = py_noiser(1920,1080)
```

```python
plt.imshow(img)
```

```python
import numpy as np
```

```python
def numpy_noiser(hres, vres):
    return np.random.randint(0,256, size = (vres, hres, 3), dtype = ('uint8')) #must be 0-256

```

```python
%timeit numpy_noiser(1920, 1080)
```

```python
img2 = numpy_noiser(1920, 1080)
```

```python
plt.imshow(img2)
```
