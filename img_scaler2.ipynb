```python
def imagescaler(image):

    '''scales an image to 50% width/height by taking a mean value of each four pixels'''
    
    
    combos = [_ for _ in itertools.product((0,1), repeat=2)] #create the four combos of 0 and 1 to get four pixels by indexing array
    new_image = np.zeros((int(image.shape[0]/2), int(image.shape[1]/2),3)) #create a new empty array for scaled image
    arr = np.zeros((int(image.shape[0]/2), int(image.shape[1]/2),4)) #create a new empty array for math operations

    for color in range(3): #for each bitplane
        for number, combo in enumerate(combos): #for each of the four combos
            arr[:,:,number] = image[combo[0]::2,combo[1]::2,color] #create four arrays stacked together
        arr_func = arr.mean(axis = 2) #take the mean of these four arrays
        new_image[:,:,color] = arr_func #this mean becomes the new scaled bitplane
    new_image = new_image.astype(int) #cast to integers
    
    return new_image
        
```

```python
plt.imshow(imagescaler(photo))
```
